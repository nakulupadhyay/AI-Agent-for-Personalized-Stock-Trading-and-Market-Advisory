```python
import yfinance as yf
import pandas as pd

def fetch_data(tickers, start="2015-01-01", end="2024-01-01"):
    data = {}
    for ticker in tickers:
        df = yf.download(ticker, start=start, end=end)
        df.dropna(inplace=True)
        data[ticker] = df
    return data
```

```python
print(fetch_data("TSLA"))
```
